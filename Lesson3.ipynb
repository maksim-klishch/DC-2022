```python
energy = pd.read_excel('Energy Indicators.xls', skiprows=17, nrows=227)
```

```python
energy.head()
```

```python
energy.tail()
```

```python
energy.drop(columns=['Unnamed: 0', 'Unnamed: 1'], inplace=True)
```

```python
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
```

```python
energy.replace('...', np.NaN, inplace=True)
```

```python
energy['Energy Supply'] = energy['Energy Supply'] * 1000000
```

```python
energy['Country'].replace({
    'Republic of Korea': 'South Korea',
    'United States of America': 'United States',
    'United Kingdom of Great Britain and Nothern Ireland': 'United Kingdom',
    'China, Hong Kong Special Administrative Region': 'Hong Kong',
}, inplace=True)
```

```python
energy['Country'] = energy['Country'].str.replace(r'([A-Za-z]*)([0-9])+', r'\1', regex=True)
```

```python
energy['Country'] = energy['Country'].str.replace(r'([A-Za-z]*) (\([A-Za-z ]*\))', r'\1', regex=True)
```
